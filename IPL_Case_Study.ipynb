```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn theme (this replaces the 'seaborn-darkgrid' confusion)
sns.set_theme(style="darkgrid")

# Optional: set color palette
sns.set_palette("Set2")
```

```python
sns.set_theme(style="whitegrid")   # Clean grid, great for bar plots
sns.set_theme(style="dark")        # Dark background
sns.set_theme(style="ticks")       # Minimalist
sns.set_theme(style="white")       # White with no grid
```

```python
top_batsmen = delivers.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top_batsmen.plot(kind='bar', title='Top 10 Batsmen by Runs', figsize=(10, 6))
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.show()
```

```python
dismissals = delivers[delivers['dismissal_kind'].isin(['bowled', 'caught', 'lbw', 'stumped', 'hit wicket'])]
top_bowlers = dismissals['bowler'].value_counts().head(10)
top_bowlers.plot(kind='bar', title='Top 10 Wicket Takers', figsize=(10, 6), color='coral')
plt.xlabel('Bowler')
plt.ylabel('Wickets')
plt.show()
```

```python
runs_per_over = delivers.groupby('over')['total_runs'].sum()
runs_per_over.plot(kind='bar', title='Total Runs per Over', figsize=(12, 6))
plt.xlabel('Over')
plt.ylabel('Total Runs')
plt.show()
```

```python
boundaries = delivers[delivers['batsman_runs'].isin([4, 6])]
boundary_counts = boundaries.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(10)
boundary_counts.plot(kind='bar', title='Top 10 Boundary Hitters', figsize=(10, 6), color='orange')
plt.xlabel('Batsman')
plt.ylabel('4s and 6s Count')
plt.show()
```

```python
delivers['dismissal_kind'].value_counts().head().plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), title='Dismissal Types')
plt.ylabel('')
plt.show()
```

```python
# Select a specific match and innings
sample_match_id = delivers['match_id'].unique()[0]  # Just pick the first match for demo

# Filter safely
innings_df = delivers[(delivers['match_id'] == sample_match_id) & (delivers['inning'] == 1)].copy()

# Check if the data is valid before proceeding
if not innings_df.empty:
    innings_df['cumulative_runs'] = innings_df['total_runs'].cumsum()
    innings_df['ball_number'] = innings_df['over'] + innings_df['ball'] / 10

    plt.figure(figsize=(10, 5))
    plt.plot(innings_df['ball_number'], innings_df['cumulative_runs'], marker='o')
    plt.title(f"Momentum Plot - Match ID {sample_match_id} (Inning 1)")
    plt.xlabel("Over.Ball")
    plt.ylabel("Cumulative Runs")
    plt.grid(True)
    plt.show()
else:
    print("No data available for the selected match and inning.")
```

```python
runs_matrix = delivers.groupby(['batsman', 'bowler'])['batsman_runs'].sum().unstack().fillna(0)
top_batsmen = runs_matrix.sum(axis=1).sort_values(ascending=False).head(5).index
top_bowlers = runs_matrix.sum().sort_values(ascending=False).head(5).index
sns.heatmap(runs_matrix.loc[top_batsmen, top_bowlers], cmap="RdYlGn", annot=True)
plt.title("Runs: Batsman vs Bowler")
plt.show()
```

```python
match_id = delivers['match_id'].unique()[0]  # You can loop over all if needed
innings_df = delivers[(delivers['match_id'] == match_id) & (delivers['inning'] == 1)]
innings_df['cumulative_runs'] = innings_df['total_runs'].cumsum()
plt.plot(innings_df['over'] + innings_df['ball']/10, innings_df['cumulative_runs'])
plt.title("Momentum Plot: Match ID {}".format(match_id))
plt.xlabel("Over.Ball")
plt.ylabel("Cumulative Runs")
plt.show()
```
